```python
import numpy as np
import pandas as pd
import glob
from functools import reduce
from ydata_profiling import ProfileReport
```

```python
files = glob.glob("/home/henrik/Documents/Thesis_master/EDA/Meas1minInterpolated/G2/*.csv")

dfs = []

for file in files:
    first_rows = pd.read_csv(file, nrows=2, header=None)
    column_name = first_rows.iloc[0, 1]

    if str(first_rows.iloc[1, 0]).strip().lower() == "unit":
        rows_to_skip = 2
    else:
        rows_to_skip = 1

    one_file_df = pd.read_csv(
        file,
        skiprows=rows_to_skip,
        header=None,
        names=["timestamp", column_name]
    )

    dfs.append(one_file_df)

df_all = reduce(lambda left, right: pd.merge(left, right, on="timestamp", how="outer"), dfs)
```

```python
print(df_all.head())
```

```python
profile = ProfileReport(df_all, title="YData Profiling Report")
profile.to_file("profiles/profile.html")
```
