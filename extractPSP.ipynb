```python
import pandas as pd

infile = "2024-01-01_2024-12-31_step_interpolation_10s/Scada.GRUN.AGG2.VANN.SJAKT.M_TRYKK.csv"
outfile = "filtered_2024_06_to_2024_11.csv"

header = open(infile).readlines()[:2]

df = pd.read_csv(infile, skiprows=2, header=None)
df[0] = pd.to_datetime(df[0])
df = df[(df[0] >= "2024-06-01") & (df[0] < "2024-10-31 00:00:00")]

open(outfile, "w").writelines(header)
df.to_csv(outfile, mode="a", index=False, header=False)
```
